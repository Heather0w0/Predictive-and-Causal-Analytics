```python
#The Answer to Q1c
import pandas as pd

damage_costs = [0, 12000, 24000, 36000, 48000, 60000, 72000, 84000]
probabilities = [0.05, 0.10, 0.15, 0.22, 0.20, 0.16, 0.09, 0.03]

# Option 1:
option1 = [42000] * len(damage_costs)

# Option 2:
# If damage <= 30000, cost = premium + damage; if damage > 30000, cost = premium + 30000.
option2 = []
premium_option2 = 15000
deductible = 30000
for d in damage_costs:
    if d <= deductible:
        cost = premium_option2 + d
    else:
        cost = premium_option2 + deductible
    option2.append(cost)

# Option 3:
option3 = damage_costs

# Create the DataFrame
payoff_matrix = pd.DataFrame({
    'Damage Cost': damage_costs,
    'Probability': probabilities,
    'Option 1 (Full Insurance)': option1,
    'Option 2 (Deductible)': option2,
    'Option 3 (Self-Insurance)': option3
})

print(payoff_matrix)
```

```python
#Answer to Q1d
import pandas as pd
import numpy as np

damage_costs = np.array([0, 12000, 24000, 36000, 48000, 60000, 72000, 84000])
probs = np.array([0.05, 0.10, 0.15, 0.22, 0.20, 0.16, 0.09, 0.03])

# Option 1: 
opt1 = np.full(len(damage_costs), 42000)

# Option 2: P
premium = 15000
deductible = 30000
opt2 = np.where(damage_costs <= deductible, premium + damage_costs, premium + deductible)

# Option 3:
opt3 = damage_costs

# Create DataFrame
df = pd.DataFrame({
    'Damage': damage_costs,
    'Probability': probs,
    'Option1': opt1,
    'Option2': opt2,
    'Option3': opt3
})

# Calculate EMV:
emv_opt1 = np.sum(df['Probability'] * df['Option1'])
emv_opt2 = np.sum(df['Probability'] * df['Option2'])
emv_opt3 = np.sum(df['Probability'] * df['Option3'])

print("EMV:")
print("Option 1:", emv_opt1)
print("Option 2:", emv_opt2)
print("Option 3:", emv_opt3)

# Build the regret matrix
min_costs = df[['Option1', 'Option2', 'Option3']].min(axis=1)
df['Regret1'] = df['Option1'] - min_costs
df['Regret2'] = df['Option2'] - min_costs
df['Regret3'] = df['Option3'] - min_costs

# Maximum regrets for each option:
max_regret_opt1 = df['Regret1'].max()
max_regret_opt2 = df['Regret2'].max()
max_regret_opt3 = df['Regret3'].max()

print("\nMax Regret:")
print("Option 1:", max_regret_opt1)
print("Option 2:", max_regret_opt2)
print("Option 3:", max_regret_opt3)

# Calculate EOL (expected opportunity loss) for each option:
eol_opt1 = np.sum(df['Probability'] * df['Regret1'])
eol_opt2 = np.sum(df['Probability'] * df['Regret2'])
eol_opt3 = np.sum(df['Probability'] * df['Regret3'])

print("\nEOL:")
print("Option 1:", eol_opt1)
print("Option 2:", eol_opt2)
print("Option 3:", eol_opt3)
```

```python
import numpy as np

projects = {
    'Alpha': np.array([23, 23, 23, 23]),
    'Bravo': np.array([17.2, 19, 22.5, 30]),
    'Charlie': np.array([12, 12, 12, 60])
}

probs = np.array([0.15, 0.15, 0.4, 0.3])

#Answer to Q2a
print("Expected Profit:")
for name, profits in projects.items():
    exp_profit = np.sum(profits * probs)
    print(f" {name}: {exp_profit:.2f}")

#Answer to Q2b
print("\nBonita's Expected Utility:")
for name, profits in projects.items():
    exp_util = np.sum(2 * np.log(profits) * probs)
    print(f" {name}: {exp_util:.4f}")

#Answer to Q2e
print("\nJasper's Expected Utility:")
for name, profits in projects.items():
    exp_util = np.sum(np.sqrt(profits) * probs)
    print(f" {name}: {exp_util:.4f}")

#Answer to Q2h
# For Bonita, solve: 2*ln(CE) = EU  => CE = exp(EU/2)
bonita_EU_charlie = np.sum(2 * np.log(projects['Charlie']) * probs)
CE_Bonita = np.exp(bonita_EU_charlie / 2)

# For Jasper, solve: sqrt(CE) = EU  => CE = (EU)^2
jasper_EU_charlie = np.sum(np.sqrt(projects['Charlie']) * probs)
CE_Jasper = jasper_EU_charlie ** 2

print("\nCertainty Equivalents for Project Charlie:")
print(f" Bonita's CE: {CE_Bonita:.2f}")
print(f" Jasper's CE: {CE_Jasper:.2f}")
```
